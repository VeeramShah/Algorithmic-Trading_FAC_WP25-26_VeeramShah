# Question 1

The first question involves arithmetic with a small application of loops. You are given an
array of n integers, that is, a1, a2, . . . , an, and you are required to output the maximum and
minimum among them.

Hint: Try with n = 3, then generalise your argument.

```python
import numpy as np

# arr = np.random.randint(1,999, (25), dtype = int)
# print(arr)
# print(arr.max())
# print(arr.min())

arr = np.array([742, 18, 593, 861, 204, 977, 36, 415, 689,
                127, 954, 301, 78, 666, 512, 49, 890, 233,
                705, 164, 928, 6, 457, 819, 290])
max = min = arr[0]
for num in arr:
  if(num>max):
    max = num
  elif(num<min):
    min = num
  else:
    continue

print(f"Maximum of array is: {max}")
print(f"Minimum of array is: {min}")
```

#Question 2
This question is for familiarity with strings. You are needed to input a string and check if it
is a palindrome.

Hint: Read a bit about the two-pointer method.

```python
str = input("Enter your input here: ")

if(str.lower() == (str[::-1]).lower()):
  print("Input is a palindrome.")
else:
  print("Input is not a palindrome.")
```

#Question 3
You will need a basic understanding of sorting for this question. Note that there is a direct
Python function to sort a sequence of integers. You are given an array of n integers, that is, a1, a2, . . . , an, and a number x. You need to find a pair of elements whose absolute difference
is x. You are assured that such a pair exists in the array.

Hint: What if you sort the array?

```python
arr = list()
for i in range(10):
  arr.append(int(input("Enter here:")))
x = int(input("Enter the unique number: "))

arr.sort()
i, j = 0, 1

while j < len(arr):
    diff = abs(arr[j] - arr[i])
    if diff == x:
        print(arr[i], arr[j])
        j += 11
    elif diff < x:
        j += 1
    else:
        i += 1
        if i == j:
            j += 1
```
